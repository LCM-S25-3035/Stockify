```python
import pandas as pd

# Download historical stock data for AAPL (Apple) from 2020-01-01 to 2023-12-31
ticker = "AAPL"
start_date = "2020-01-01"
end_date = "2023-12-31"

df_stock = yf.download(ticker, start=start_date, end=end_date)

# Reset index to have 'Date' as a column (similar to your news data)
df_stock.reset_index(inplace=True)

# Preview stock data columns and first few rows
print("Columns:", df_stock.columns)
print(df_stock.head())

#save the data in csv
df_stock.to_csv('aapl_stock_data.csv', index=False)
print("\n✅ Stock data saved to 'aapl_stock_data.csv'")
```

```python
# prompt: visualize the above data

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_stock['Date'], df_stock['Close'])
plt.title(f'{ticker} Close Price over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```
